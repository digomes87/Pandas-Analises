```python
import pandas as pd
import numpy as np
```

```python
df_obsety = pd.read_csv('/workspaces/Pandas-Analises/Obsidade/obesity_cleaned.csv')
df_obsety
```

```python
# 1. Limpeza dos dados
# Extrai o valor numérico antes do espaço na coluna "Obesity (%)"
df_obsety['Obesity'] = df_obsety['Obesity (%)'].str.extract(r'([\d.]+)').astype(float)
df_obsety
```

```python
# 2. Cálculo do percentual médio de obesidade por sexo no mundo em 2015
# Filtra apenas o ano de 2015 e os sexos "Male" e "Female"
df_2015 = df_obsety[(df_obsety['Year'] == 2015) & (df_obsety['Sex'].isin(['Male', 'Female']))]
df_2015 = df_2015.sort_values(by='Obesity', ascending=False)
df_2015
```

```python
# Agrupa por sexo e calcula a média de obesidade
obesity_by_sex_2015 = df_2015.groupby('Sex')['Obesity'].mean().reset_index()
obesity_by_sex_2015
```
